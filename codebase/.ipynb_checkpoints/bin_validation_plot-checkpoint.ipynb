```python
# notebook import

from ipynb.fs.full.data_processing import *
```

```python
# user input parameters

# boolean parameter to dictate whether figures are saved
save = False 

# list of timesteps for which to calculate mutual information for all subsets (should match `bin_validation.ipynb`)
timesteps = [1000, 2000, 3000, 4000, 5000]

# list of numbers over which the plot of mi will be averaged 
mi_avg_list = [0, 10, 20]

# list of colors used in figures
colors = ['lightcoral', 'orange', 'yellowgreen', 'limegreen', 'deepskyblue', 'slateblue', 'orchid', 'pink']
```

```python
# input: list of lists of subset sizes (x_timsteps), list of lists of mutual information measurements (y_timesteps), and a list of their corresponding feature and average (feature, avg) 
def plot_validation_mi(x_timsteps, y_timesteps, feature, avg):
    for (x, y, t, c) in zip(x_timsteps, y_timesteps, timesteps, colors):
        plt.plot(x, y, label=t, color=c)
        
    plt.title(f'{proper_feature[feature]} mutual information as a function of subset size')
    plt.xlabel('Subset Size')
    plt.ylabel('Mutual Information')
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

    if save:
        plt.savefig(results_fpath + bin_validation_ext + f'/validation_mi_plots/{feature}_validation_mi_{avg}.png', dpi=300,  bbox_inches='tight')
    
    plt.show()
    return
```

```python
# `main()` function declaration

# intended use in coordination with `bin_validation.ipynb`
# requires: import of `data_processing.ipynb` and user input parameters specified above
# output: None (plots the validation mi)

def main():
    
    # list of timesteps relative to input data
    global unscaled
    unscaled = get_timesteps_unscaled(timesteps)
    
    # read validation dataframe
    validation_mi_df = pd.read_csv(results_fpath + bin_validation_ext + 'validation_mi.csv')
    
    for feature in feature_name_list:
        
        feature_df = validation_mi_df.loc[validation_mi_df['feature'] == f'{feature}']
        
        for avg in mi_avg_list:
            
            x_timesteps = []
            y_timesteps = []
        
            for t in unscaled:

                feature_t_df = feature_df.loc[feature_df['timestep'] == t]

                x_mi, y_mi = avg_plot(list(feature_t_df['subset_size']), list(feature_t_df['mi']), avg)
                
                x_timesteps.append(x_mi)
                y_timesteps.append(y_mi)
                
                plot_validation_mi(x_timesteps, y_timesteps, feature, avg)
    
    return         
```

```python
# call to `main()` function

main()
```
