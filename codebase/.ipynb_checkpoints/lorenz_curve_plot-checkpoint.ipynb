```python
# notebook import 

from ipynb.fs.full.data_processing import *
```

```python
# user input parameters

# boolean parameter to dictate whether figures are saved
save = True

# interval for times of matrices
interval = 1000

# dictionary mapping each plotted line to its corresponding color in figures
color_dict = {'gfp':'green', 'rfp':'red', 'area':'blue'}

# list of colors corresponding to specific wells used in figures
colors = ['sienna', 'tomato', 'orange', 'darkgoldenrod', 'yellowgreen', 'limegreen', 'teal', 'deepskyblue', 'royalblue', 'slateblue', 'darkviolet', 'orchid', 'pink']
```

```python
# function declaration

# input: None (references the `interval` global input parameter)
# output: list of timesteps (times), and intervals at which they were gathered (intervals)
def get_times():
    times = []
    intervals = []
    for i in range(len(timestep_scaled)):
        if timestep_scaled[i] >= interval*(len(times)+1):
            times.append(i)
            intervals.append(interval*(len(times)))
    return times, intervals

# input: series of measurements
# output: None (plots the Lorenz curve of the distribution)
def plot_lorenz_curve(val_list, feature, color, t, well=None):
    x = val_list.cumsum() / val_list.sum()
    x = np.insert(x, 0, 0) 
    x[0], x[-1]
   
    plt.scatter(np.arange(x.size)/(x.size-1), x, 1, color=color)
    plt.plot([0,1], [0,1], color='grey')
    
    if well == None:
         plt.title(f'Combined {proper_feature[feature]} Lorenz plot ({t} min)')
    else:
        plt.title(f'{proper_feature[feature]} Lorenz plot ({well} at {t} min)')
    plt.xlabel('Cumulative Percentage of Population')
    plt.ylabel(f'Cumulative Percentage of {proper_feature[feature]}')
    
    if save: 
        if well == None:
            plt.savefig(results_fpath + gini_ext + f'/lorenz_curve_plots/{feature}_lorenz_combined_{t}.png', dpi=300,  bbox_inches='tight')
        else:
            plt.savefig(results_fpath + gini_ext + f'/lorenz_curve_plots/{feature}_lorenz_{well}_{t}.png', dpi=300,  bbox_inches='tight')
        
    plt.show()
    return
```

```python
def main():
    
    # get the times and intervals for which the plots will be made
    times, intervals = get_times()
    
    all_df = create_df_subset(experiment_dict, ['mean_F_C2', 'mean_F_C3', timestep_ref], ['gfp', 'rfp', 'time'])

    for name, ref in zip(feature_name_list, feature_ref_list):
        
        if name != 'area':

            for (t, i) in zip(times, intervals):
                
                plt.plot([0,1], [0,1], color='grey')
                
                for well, color in zip(well_id_list, colors):
                    
                    val_list = experiment_dict[well].loc[experiment_dict[well][timestep_ref] == t][ref]
                    val_list = np.array(val_list.sort_values())
                    
                    x = val_list.cumsum() / val_list.sum()
                    x = np.insert(x, 0, 0) 
                    x[0], x[-1]
   
                    plt.scatter(np.arange(x.size)/(x.size-1), x, 2, color=color, alpha = 0.5, label=well)
                
                plt.title(f'{proper_feature[name]} Lorenz plot ({i} min)')
                plt.xlabel('Cumulative Percentage of Population')
                plt.ylabel(f'Cumulative Percentage of {proper_feature[name]}')
                plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
                
                if save: 
                    plt.savefig(results_fpath + gini_ext + f'/lorenz_curve_plots/{name}_lorenz_all_{i}.png', dpi=300,  bbox_inches='tight')
                
                plt.show()
                    
            for (t, i) in zip(times, intervals):
                
                val_list = all_df.loc[all_df['time'] == t][name]
                plot_lorenz_curve(np.array(val_list.sort_values()), name, color_dict[name], i)
        
```

```python
main()
```
