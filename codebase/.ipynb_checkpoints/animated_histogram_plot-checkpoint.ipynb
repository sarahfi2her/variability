```python
# notebook import

from ipynb.fs.full.data_processing import *
```

```python
# list of colors corresponding to specific wells used in figures
colors = ['sienna', 'tomato', 'orange', 'darkgoldenrod', 'yellowgreen', 'limegreen', 'teal', 'deepskyblue', 'royalblue', 'slateblue', 'darkviolet', 'orchid', 'pink']
```

```python
def prepare_animation(bar_container, well, ref, bins, color):

    def animate(t):
        # simulate new data coming in
        data = (experiment_dict[well].loc[experiment_dict[well][timestep_ref] == (t+1)*div])[ref]
        n, _ = np.histogram(data, bins)
        for count, rect in zip(n, bar_container.patches):
            rect.set_height(count)
        return bar_container.patches
    return animate

```

```python
def main():
    
    global div
    div = 1
    
    global bin_edges_dict
    bin_edges_dict = create_bin_edges_dict()
    
    for name, ref in zip(feature_name_list, feature_ref_list):
        if name != 'area':
            
            bins = bin_edges_dict[name]
            
            for well, color in zip(well_id_list, colors):
                print(f'doing well {well} !')
                data = (experiment_dict[well].loc[experiment_dict[well][timestep_ref] == 1])[ref]
                n, _ = np.histogram(data, bins)
                
                fig, ax = plt.subplots()
                _, _, bar_container = ax.hist(data, gfp_bins, lw=1, ec=color, fc=color)
                
                
                if name == 'gfp':
                    ax.set_ylim(top=100)  # set safe limit to ensure that all data is visible.
                elif name == 'rfp':
                    ax.set_ylim(top=200)
                
                plt.title(f'{proper_feature[name]} distribution through time ({well})')
                plt.xlabel(f'{proper_feature[name]}')
                plt.ylabel('Number of Cells')
                print(f'ready to animate !')
                ani = animation.FuncAnimation(fig, prepare_animation(bar_container, well, ref, bins, color), len(timestep_list)//div, repeat=False, blit=True)
                print(f'done animating !')
                #video = ani.to_html5_video()
    
                # embedding for the video
                #html = display.HTML(video)

                # draw the animation
                #display.display(html)

                writervideo = animation.FFMpegWriter(fps=60)
                ani.save(results_fpath + f'animated_histograms/{name}_{well}.mp4', dpi=300, writer=writervideo)
                #ani.save(results_fpath + f'animated_histograms/{name}_{well}.mp4', dpi=300)
                print(f'that was a big video !')
                plt.close()
                
```

```python
main()
```
